```python
X = datos_na.drop(columns= ["median_house_value"],axis=1)
y = datos_na["median_house_value"]
```

```python
from sklearn.model_selection import train_test_split

x_ent, x_prue, y_ent, y_prue = train_test_split(X,y,test_size = .2)
```

```python
from sklearn.linear_model import LinearRegression

modelo = LinearRegression()
```

```python
modelo.fit(x_ent,y_ent)
```

```python
predicciones = modelo.predict(x_prue)
```

```python
predicciones
```

```python
comparativa = {"Prediccion": predicciones, "Valor Real": y_prue }
pd.DataFrame(comparativa)
```

```python
# overfitting
print(modelo.score(x_ent,y_ent))
print(modelo.score(x_prue,y_prue))
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

mse  = mean_squared_error(y_prue,predicciones)
```

```python
mse
```

```python
rsme = np.sqrt(mse)
```
