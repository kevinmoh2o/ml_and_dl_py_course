```python
import pandas as pd

datos = pd.read_csv("../lab01/celsius.csv")
```

```python
x = datos["Celsius"].values
y = datos["Fahrenheit"].values
```

```python
datos.head()
```

```python
import tensorflow as tf

modelo = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(1,)),
    tf.keras.layers.Dense(units=1),
])

modelo.compile(
    optimizer = tf.keras.optimizers.Adam(.1),
    loss= "mean_squared_error"
)
```

```python
entrenamiento = modelo.fit(x,y,epochs=1000)
```

```python
import matplotlib.pyplot as plt


plt.xlabel("# Epoca")
plt.ylabel("MAgnitud perdida")
plt.plot(entrenamiento.history["loss"])

```

```python
import numpy as np

c = 123
sentrada = np.array([[c]], dtype=np.float32)

prediccion = modelo.predict(entrada)
print(f"Predicción: {c} Celsius son {prediccion[0][0]} Fahrenheit")
```
